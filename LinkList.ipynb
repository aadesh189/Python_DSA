# **Linked List**
Its A linear data strcture, where elements are stored in a sequence.

Unlike array or list, a link list does not store elements in contiguous memory.

### Each node contain two parts
1.   Data - The actual data or value that the node holds.
2.   Next - A referance to the next node in the sequence.

The first node in the linked list is reffered as the head, and last node points to None.

### Characteristics of linked list

1.   Dynamic size - Linked list do not required a pre-defined size,  unlike array. They can grow or shrink dynamically as elements are added or removed.
2.   Non-contiguous memory allocation - Nodes are not stored in contiguous memory locations. Each node points to the next node in the list.

### Why use linked list

1.   Efficient insertion/ Deletion - Link list are more efficient than arrays when it comes to inserting and deleting elements, especially when dealing with large dataset.

Insertion and deletion at the begining or middle of the linked list can be done in constant time O(n)
2.   Dynamic size - Unlike arrays, linked list can grow or shrik dynamically without needing to reallocate or recoganize memory.

### Types of linked list.

1. **Single linnked list** - Each node points only to the next node.

2. **Circular linked list** - The last node points back to the first node, forming a circular chain.

3. **Doubly linked list** - Each node has two referance, one pointing to the next node and one pointing to the previous node.

Creating Link list

```python
# Creating link list
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None

#Creating nodes
node1=Node(10)
node2=Node(20)
node3=Node(30)
node4=Node(40)
node5=Node(50)

#Connecting nodes to form a linked list
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

#Printing the linked list
current = node1
while current is not None:
  print(current.data, end="->")
  current=current.next
print("None")
```
